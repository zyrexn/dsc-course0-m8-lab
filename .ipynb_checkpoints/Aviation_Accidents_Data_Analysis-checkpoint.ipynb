```python
# loading relevant packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Exploratory Data Analysis  
- Load in the cleaned data

```python
# Load the cleaned dataset from Part 1
df = pd.read_csv('data/cleaned_aviation_data.csv', low_memory=False)

# Inspect the first few rows
df.head()
```

```python
# Split into small and large planes using 20 passengers as threshold
small_planes = df[df['Total.Passengers'] <= 20]
large_planes = df[df['Total.Passengers'] > 20]

# Check the sizes
print(f"Small planes: {small_planes.shape[0]} rows")
print(f"Large planes: {large_planes.shape[0]} rows")
```

```python
# Get the 15 small makes with lowest mean fatal/serious injury fraction
small_makes = (small_planes.groupby('Make')['Fatal.Serious.Fraction']
               .mean()
               .dropna()
               .sort_values()
               .head(15))

# Get the 15 large makes with lowest mean fatal/serious injury fraction
large_makes = (large_planes.groupby('Make')['Fatal.Serious.Fraction']
               .mean()
               .dropna()
               .sort_values()
               .head(15))

print("Top 15 safest small plane makes:")
print(small_makes)
print('---')
print("Top 15 safest large plane makes:")
print(large_makes)
```

```python
# Plot mean fatal/serious injury fraction for small and large makes side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Small planes plot
small_makes.plot(kind='barh', ax=axes[0], color='steelblue')
axes[0].set_title('Top 15 Safest Small Plane Makes')
axes[0].set_xlabel('Mean Fatal/Serious Injury Fraction')
axes[0].set_ylabel('Make')

# Large planes plot
large_makes.plot(kind='barh', ax=axes[1], color='coral')
axes[1].set_title('Top 15 Safest Large Plane Makes')
axes[1].set_xlabel('Mean Fatal/Serious Injury Fraction')
axes[1].set_ylabel('Make')

plt.tight_layout()
plt.show()
```

```python
# Get top 10 safest small makes for violin plot
top10_small = small_makes.head(10).index

# Filter small planes for those makes
small_top10_df = small_planes[small_planes['Make'].isin(top10_small)]

# Violin plot for small plane injury rate distribution
plt.figure(figsize=(14, 6))
sns.violinplot(data=small_top10_df, x='Make', y='Fatal.Serious.Fraction')
plt.title('Distribution of Injury Rates - Top 10 Safest Small Plane Makes')
plt.xlabel('Make')
plt.ylabel('Fatal/Serious Injury Fraction')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Get top 10 safest large makes for strip plot
top10_large = large_makes.head(10).index

# Filter large planes for those makes
large_top10_df = large_planes[large_planes['Make'].isin(top10_large)]

# Strip plot for large plane injury rate distribution
plt.figure(figsize=(14, 6))
sns.stripplot(data=large_top10_df, x='Make', y='Fatal.Serious.Fraction')
plt.title('Distribution of Injury Rates - Top 10 Safest Large Plane Makes')
plt.xlabel('Make')
plt.ylabel('Fatal/Serious Injury Fraction')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Destruction rate by make for small planes
small_destroyed = (small_planes.groupby('Make')['Destroyed']
                   .mean()
                   .sort_values()
                   .head(15))

# Destruction rate by make for large planes
large_destroyed = (large_planes.groupby('Make')['Destroyed']
                   .mean()
                   .sort_values()
                   .head(15))

print("Small planes - lowest destruction rates:")
print(small_destroyed)
print('---')
print("Large planes - lowest destruction rates:")
print(large_destroyed)
```

## Make Recommendations

### Small Planes
Based on the analysis of fatal/serious injury fraction and destruction rates, 
the top recommended small plane makes are:
- **Bombardier Inc** and **Mcdonnell Douglas**: lowest mean injury fractions (~0.09-0.09)
- **Maule** and **Aviat Aircraft Inc**: consistently low injury and destruction rates
- **Boeing** and **Aeronca**: low destruction rates with reasonable injury fractions

### Large Planes
For large planes, the standout makes are:
- **Cessna, Dehavilland, Grumman, Piper**: 0% destruction rate
- **Mcdonnell Douglas** and **Bombardier Inc**: near-zero injury fractions
- **Boeing** and **Embraer**: low destruction rates with low injury fractions
- Note: **Beech** performs poorly with a 100% destruction rate and should be avoided

```python
# Filter models with at least 10 examples
model_counts = df.groupby('Make_Model')['Fatal.Serious.Fraction'].count()
valid_models = model_counts[model_counts >= 10].index
df_models = df[df['Make_Model'].isin(valid_models)]

# Split into small and large
small_models = df_models[df_models['Total.Passengers'] <= 20]
large_models = df_models[df_models['Total.Passengers'] > 20]

# Get top 10 safest small models
top10_small_models = (small_models.groupby('Make_Model')['Fatal.Serious.Fraction']
                      .mean()
                      .sort_values()
                      .head(10))

print("Top 10 safest small plane models:")
print(top10_small_models)
```

```python
# Plot top 10 safest small models
plt.figure(figsize=(12, 6))
top10_small_models.plot(kind='barh', color='steelblue')
plt.title('Top 10 Safest Small Plane Models - Mean Injury Fraction')
plt.xlabel('Mean Fatal/Serious Injury Fraction')
plt.ylabel('Model')
plt.tight_layout()
plt.show()

# Get top 10 safest large models
top10_large_models = (large_models.groupby('Make_Model')['Fatal.Serious.Fraction']
                      .mean()
                      .sort_values()
                      .head(10))

print("Top 10 safest large plane models:")
print(top10_large_models)
```

```python
# Plot top 10 safest large models
plt.figure(figsize=(12, 6))
top10_large_models.plot(kind='barh', color='coral')
plt.title('Top 10 Safest Large Plane Models - Mean Injury Fraction')
plt.xlabel('Mean Fatal/Serious Injury Fraction')
plt.ylabel('Model')
plt.tight_layout()
plt.show()
```

```python
# Factor 1: Weather Condition vs Injury Fraction
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Weather.Condition', y='Fatal.Serious.Fraction')
plt.title('Injury Fraction by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Fatal/Serious Injury Fraction')
plt.tight_layout()
plt.show()

# Summary statistics
print(df.groupby('Weather.Condition')['Fatal.Serious.Fraction'].describe())
```

```python
# Factor 2: Engine Type vs Injury Fraction
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Engine.Type', y='Fatal.Serious.Fraction')
plt.title('Injury Fraction by Engine Type')
plt.xlabel('Engine Type')
plt.ylabel('Fatal/Serious Injury Fraction')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Summary statistics
print(df.groupby('Engine.Type')['Fatal.Serious.Fraction'].describe())
```

## Factor Analysis

### Factor 1: Weather Condition
Weather condition has a strong effect on injury outcomes:
- **IMC** (bad weather/low visibility): mean injury fraction of 0.63 — nearly 3x higher than VMC
- **VMC** (good weather/clear visibility): mean injury fraction of 0.23
- This strongly suggests that flying in poor weather conditions significantly 
increases the risk of fatal or serious injuries in the event of an accident.

### Factor 2: Engine Type
Engine type also affects safety outcomes:
- **Turbo Fan** engines have the lowest mean injury fraction at 0.09 — 
  these are typical of large commercial jets
- **Reciprocating** engines (common in small planes) have a mean of 0.25
- **Turbo Prop** engines have the highest mean at 0.32
- This suggests that jet-powered aircraft (Turbo Fan) tend to be safer 
  in accident scenarios, likely due to better engineering and redundancy systems
